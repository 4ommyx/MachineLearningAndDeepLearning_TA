```python
## Matplotlib
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 1000)
r = 9
x = r * np.cos(theta)
y = r * np.sin(theta)

regions = [
    (0, np.pi/2, 'red'),       # ซีกขวาบน
    (np.pi/2, np.pi, 'purple'),      # ซีกซ้ายบน
    (np.pi, 3*np.pi/2, 'pink'),  # ซีกซ้ายล่าง
    (3*np.pi/2, 2*np.pi, 'skyblue')    # ซีกขวาล่าง
]

plt.figure(figsize=(6, 6))
for start, end, color in regions:
    mask = (theta >= start) & (theta < end)
    plt.plot(x[mask], y[mask], color=color, linewidth=2)

plt.title("I love Matplotlib & P'TA MAK MAK")
```

```python
## Pandas
import kagglehub
import os
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("shivamb/netflix-shows")

print("Path to dataset files:", path)
print(os.listdir(path))
```

```python
file_path = os.path.join(path, os.listdir(path)[0])
df = pd.read_csv(file_path)

df.head()
```

```python
missing_counts = df.isnull().sum()
missing_counts
```

```python
missing_counts.plot.bar()
```

```python
cols_to_dropna = missing_counts[missing_counts > 800].index
cols_to_dropna
```

```python
df_cleaned = df.dropna(subset=cols_to_dropna)
missing_counts_af = df_cleaned.isnull().sum()
missing_counts_af
```

```python
missing_counts_af.plot.bar()
```
